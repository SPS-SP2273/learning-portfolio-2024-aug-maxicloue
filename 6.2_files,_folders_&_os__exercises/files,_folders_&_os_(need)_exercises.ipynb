## Exercise 1 (Tidying a collaboration)

```python
import os, glob, shutil
```

### Task 1

```python
original = 'oct-08_singapore.txt'
dictionary = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}

parts = original.split('-')
month = parts [0]
parts2 = parts[1].split('_')
date = parts2[0]
city = parts2[1]
month_number = dictionary.get(month)
new = f'{month_number}-{date}-{city}'
print (new)
```

### Task 2

```python
def rename_my_file(old_file_name):
    dictionary = {
    'jan': '01', 'feb': '02', 'mar': '03',
    'apr': '04', 'may': '05', 'jun': '06',
    'jul': '07', 'aug': '08', 'sep': '09',
    'oct': '10', 'nov': '11', 'dec': '12'
    }

    parts = old_file_name.split('-')
    month = parts [0]
    parts2 = parts[1].split('_')
    date = parts2[0]
    city = parts2[1]
    month_number = dictionary.get(month)
    new_file_name = f"{month_number}-{date}-{city}"
    return new_file_name

print(rename_my_file('oct-08_singapore.txt'))
```

### Task 3

```python
dates = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
for day in dates:
    print(f'{day} ----> {rename_my_file(day)}')
```

### Task 4

```python
for files in glob.glob('**/*.txt'):
    cleaned_up = files.split(os.path.sep)[1]
    print (rename_my_file(cleaned_up))
```

### Task 5

```python
Cities = ['cairo', 'dhaka', 'jakarta', 'karachi', 'manila',
 'melbourne', 'osaka', 'shanghai', 'singapore', 'tokyo']
os.mkdir('City')
for city in Cities:
    path = os.path.join('City', city)
    print (f'Creating {path}')
    os.mkdir(path)
```

### Task 6

```python
glob.glob('**/*cairo.txt')
```

```python
for files in glob.glob('**/*cairo.txt'):
    path_to_destination = os.path.join('City', 'cairo')
    shutil.copy(files, path_to_destination)
    print(f'Copied file to {path_to_destination}')
```

### Task 7

```python
for city in Cities:
    for file in glob.glob(f'*/*{city}.txt'):
        path_desti = os.path.join('City', f"{city}")
        if os.path.exists(os.path.join(path_desti, file)):
            continue
        
        shutil.copy(file, path_desti)
        print (f'Copied file to {path_desti}')
```
